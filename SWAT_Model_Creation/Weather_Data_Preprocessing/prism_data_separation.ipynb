# Inputs

```python
# Replace filepath with PRISM raw data CSV (keep the r in front)
prism_rawdata = pd.read_csv(r"C:\ExampleFilepath\Raw Data\PRISM_ppt_tmin_tmax_stable_4km_20100101_20101231.csv", header=10)
```

```python
# Set the number of locations (determined during PRISM data collection)
n_locations = 100
```

```python
# Set the filepath/filename of the output precipitation CSV (keep the r in front). Include data year in name
# See Weather Data Protocol for recommended folder organization (Intermediate Outputs --> Precipitation)
precipitation_csv = r"C:\ExampleFilepath\Intermediate Outputs\Precipitation\precipitation_2010.csv"
```

```python
# Set the filepath/filename of the output minimum temperature CSV (keep the r in front). Include data year in name.
# See Weather Data Protocol for recommended folder organization (Intermediate Outputs --> Temperature --> Minimum Temperature)
mintemp_csv = r"C:\ExampleFilepath\Intermediate Outputs\Temperature\Minimum Temperature\mintemp_2010.csv"
```

```python
# Set the filepath/filename of the output maximum temperature CSV (keep the r in front). Include data year in name.
# See Weather Data Protocol for recommended folder organization (Intermediate Outputs --> Temperature --> Maximum Temperature)
maxtemp_csv = r"C:\ExampleFilepath\Intermediate Outputs\Temperature\Maximum Temperature\maxtemp_2010.csv"
```

# Code

```python
# To be used for the date column
date_prism = prism_rawdata['Date'][prism_rawdata['Name'] == 'Location 1']
```

## Precipitation Extraction

```python
# Creates dataframe with columns of daily precipitation data for each location and then writes the dataframe to a CSV

for i in range(1, n_locations + 1):
    globals()[f'pdf_{i}'] = prism_rawdata.loc[
        prism_rawdata['Name'] == f'Location {i}',
        'ppt (mm)'
    ].reset_index(drop=True)

Precipitation_Prism_data = pd.DataFrame(
    {
        'date_prism': date_prism,
        **{f'pdf_{i}': globals()[f'pdf_{i}']
           for i in range(1, n_locations + 1)}
    }
)

Precipitation_Prism_data.to_csv(precipitation_csv, index=False)
```

## Minimum Temperature Extraction 

```python
# Creates dataframe with columns of daily minimum temperature data for each location and then writes the dataframe to a CSV

for i in range(1, n_locations + 1):
    globals()[f'mntdf_{i}'] = prism_rawdata.loc[
        prism_rawdata['Name'] == f'Location {i}',
        'tmin (degrees C)'
    ].reset_index(drop=True)

Minimum_Temperature_Prism_data = pd.DataFrame(
    {
        'date_prism': date_prism,
        **{f'mnt_df_{i}': globals()[f'mnt_df_{i}']
           for i in range(1, n_locations + 1)}
    }
)

Minimum_Temperature_Prism_data.to_csv(mintemp_csv, index=False)
```

## Maximum Temperature Extraction

```python
# Creates dataframe with columns of daily maximum temperature data for each location and then writes the dataframe to a CSV

for i in range(1, n_locations + 1):
    globals()[f'mxtdf_{i}'] = prism_rawdata.loc[
        prism_rawdata['Name'] == f'Location {i}',
        'tmax (degrees C)'
    ].reset_index(drop=True)

Maximum_Temperature_Prism_data = pd.DataFrame(
    {
        'date_prism': date_prism,
        **{f'mxtdf_{i}': globals()[f'mxtdf_{i}']
           for i in range(1, n_locations + 1)}
    }
)

Maximum_Temperature_Prism_data.to_csv(maxtemp_csv, index=False)
```
