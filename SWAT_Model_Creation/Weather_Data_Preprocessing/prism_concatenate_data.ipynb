# Inputs

```python
# Replace filepath with input Precipitation folder location (keep the r in front)
# See Weather Data Protocol for recommended folder organization (Intermediate Outputs --> Precipitation)
precipitation_folder = r"C:\ExampleFilepath\Intermediate Outputs\Precipitation"
```

```python
# Replace filepath with input Minimum Temperature folder location (keep the r in front)
# See Weather Data Protocol for recommended folder organization (Intermediate Outputs --> Temperature --> Minimum Temperature)
mintemp_folder = r"C:\ExampleFilepath\Intermediate Outputs\Temperature\Minimum Temperature"
```

```python
# Replace filepath with input Maximum Temperature folder location (keep the r in front)
# See Weather Data Protocol for recommended folder organization (Intermediate Outputs --> Temperature --> Maximum Temperature)
maxtemp_folder = r"C:\ExampleFilepath\Intermediate Outputs\Temperature\Maximum Temperature"
```

```python
# Replace filepath with folder that will contain output concatenated/merged CSV files (all three output CSVs can be kept in a single folder)
# See Weather Data Protocol for recommended folder organization (Intermediate Outputs --> MergedCSVs)
mergedcsvs_folder = r"C:\ExampleFilepath\Intermediate Outputs\MergedCSVs"
```

## Precipitation Data Concatenation

```python
# Create list of csv filepaths that are in provided Precipitation folder
precipitation_filepaths = [
    os.path.join(precipitation_folder, f)
    for f in os.listdir(precipitation_folder)
    if f.lower().endswith(".csv")
    and os.path.isfile(os.path.join(precipitation_folder, f))
]
```

```python
# Reads all provided CSV files for precipitation and puts the dataframes in a list
pdf_list = [pd.read_csv(file) for file in precipitation_filepaths]

# Concatenates dataframe list into a single precipitation dataframe
precip_df = pd.concat(pdf_list, ignore_index=True)
```

```python
# Save the merged dataframe to a new CSV file to be stored in the output folder defined in the Inputs section
merged_precip_filepath = os.path.join(mergedcsvs_folder, "Precipitation_merged.csv")
precip_df.to_csv(merged_precip_filepath, index=False)
```
